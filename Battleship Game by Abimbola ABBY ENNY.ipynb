```python
import random

#To print Board
def print_board(board):
    for row in board:
        print(" ".join(row))

def random_row_col(board):
    return random.randint(0, len(board) - 1), random.randint(0, len(board[0]) - 1)

##return random.randint(0, len(board) - 1), generates a random row index between 0 and the number of rows in the board - 1

#random.randint(0, len(board[0]) - 1) , generates a random column index between 0 and the number of columns in the board - 1


def is_valid_move(row, col, board):
    return 0 <= row < len(board) and 0 <= col < len(board[0])

##This part checks if the row/column index is greater than or equal to 0 and less than the number of rows/column in the board
# it ensures that the row/column index is within the valid range of rows/column on the board.

def is_ship_sunk(ship, hits):
    return all(board[row][col] == "X" for row, col in ship)

def battleship_game():
    print('Welcome to Battleship!')
    print("This is a double player game")
    # Set up the board
    board_size = 5
    board = [["O" for _ in range(board_size)] for _ in range(board_size)]

    # Place ships for each player
    ships_per_player = 1
    player_1_ships = [random_row_col(board) for _ in range(ships_per_player)]
    player_2_ships = [random_row_col(board) for _ in range(ships_per_player)]

    players = [player_1_ships, player_2_ships]

    # Game loop
    current_player = 0

    while True:
        print("\nPlayer {}'s Turn".format(current_player + 1))
        print_board(board) ##this will print player 1's Turn ,since current player = 0

        # Take user input for a guess
        while True:
            try:
                row = int(input("Enter the row (0-{}): ".format(board_size - 1)))
                col = int(input("Enter the column (0-{}): ".format(board_size - 1)))
                break
            except ValueError:
                print("Invalid input. Please enter a number.")

        # Check if the move is valid
        if not is_valid_move(row, col, board):
            print("Invalid move. Try again.")
            continue

        # Check if the move hits a ship
        if (row, col) in players[1 - current_player]:
            print("Hit!")
            players[1 - current_player].remove((row, col))
            board[row][col] = "X"
        else:
            print("Miss!")
            board[row][col] = "M"

        # Check if the player has sunk all the opponent's ships
        if not players[1 - current_player]:
            print("\nPlayer {} wins!".format(current_player + 1))
            break

        # Switch to the other player
        #Assume current_player is currently 0.
#1 - current_player calculates 1 - 0, resulting in 1.
#Now, current_player is set to 1.

#On the next turn:

#Assume current_player is currently 1.
#1 - current_player calculates 1 - 1, resulting in 0.
#Now, current_player is set back to 0

        current_player = 1 - current_player

if __name__ == "__main__":
    battleship_game()
```
